```python
# 1.1 HW1 - ICS4Ufactorial.py
def factorial(N):
    if N == 0:
        return 1
    else:
        return N * factorial(N-1)
```

```python
# 1.1 HW2 - ICS4Usquared.py
def squared(N):
    if N == 1:
        return 1
    else:
        return squared(N-1) + 2 * N - 1
```

```python
# 1.1 HW3 - ICS4UpyramidalNumbers.py
def triangle(N):
    if N == 1:
        return 1
    else:
        return N + triangle(N-1)

def pyramidal(N):
    if N == 1:
        return 1
    else:
        return pyramidal(N-1) + triangle(N)

def main():
    num = int(input("Enter N: "))
    print(pyramidal(num))

main()
```

```python
# 1.1 HW4 - ICS4UsumOfList.py
def sumList(theList,index):
    if index == len(theList):
        return 0
    else:
        return theList[index] + sumList(theList, index + 1)


def main():
    x = [1,2,3,4,5]
    print(sumList(x,0))

    x = [4,3,7,12,17,20,22]
    print(sumList(x, 0))


main()
```
